```python
import pandas as pd
import numpy as np
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

# Agrupamento dos dados do ENEM

```python
year = '2021'
```

```python
df_enem = pd.read_csv(os.getenv('PATH_TO_RAW_DATA')+'MICRODADOS_ENEM_'+year+'.csv', encoding = "ISO-8859-1", delimiter=';')
df_enem.head()
```

```python
df_enem.columns
```

```python
df_enem = df_enem[['NU_INSCRICAO', 'NU_ANO','TP_COR_RACA','TP_ESCOLA', 'TP_ENSINO','SG_UF_ESC','NO_MUNICIPIO_ESC','TP_DEPENDENCIA_ADM_ESC', 'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC','NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO','Q001', 'Q002','Q006','Q025']]
df_enem.head()
```

```python
df_enem = df_enem.rename(columns={'Q001': 'ESCOLARIDADE_PAI', 'Q002': 'ESCOLARIDADE_MAE', 'Q006': 'RENDA_MENSAL', 'Q025': 'INTERNET'})
```

```python
replace_TP_COR_RACA = {0: 'nao_declarado',
                     1: 'branca',
                     2: 'preta',
                     3: 'parda',
                     4: 'amarela',
                     5: 'indigena',
                     6: 'nao_dispoe_informacao'}

df_enem['TP_COR_RACA'] = df_enem['TP_COR_RACA'].replace(replace_TP_COR_RACA)
```

```python
replace_TP_ESCOLA = {1: 'nao_respondeu',
                     2: 'publica',
                     3: 'privada'}

df_enem['TP_ESCOLA'] = df_enem['TP_ESCOLA'].replace(replace_TP_ESCOLA)
```

```python
replace_TP_ENSINO = {1: 'ensino_regular',
                     2: 'educacao_especial'}

df_enem['TP_ENSINO'] = df_enem['TP_ENSINO'].replace(replace_TP_ENSINO)
```

```python
replace_TP_DEPENDENCIA_ADM_ESC = {1: 'federal',
                                  2: 'estadual',
                                  3: 'municipal',
                                  4: 'privada'}

df_enem['TP_DEPENDENCIA_ADM_ESC'] = df_enem['TP_DEPENDENCIA_ADM_ESC'].replace(replace_TP_DEPENDENCIA_ADM_ESC)
```

```python
replace_TP_LOCALIZACAO_ESC = {1: 'urbana',
                              2: 'rural'}

df_enem['TP_LOCALIZACAO_ESC'] = df_enem['TP_LOCALIZACAO_ESC'].replace(replace_TP_LOCALIZACAO_ESC)
```

```python
replace_TP_SIT_FUNC_ESC = {1: 'em_atividade',
                           2: 'paralisada',
                           3: 'extinta',
                           4: 'extinta_em_anos_anteriores'}

df_enem['TP_SIT_FUNC_ESC'] = df_enem['TP_SIT_FUNC_ESC'].replace(replace_TP_SIT_FUNC_ESC)
```

```python
replace_INTERNET = {'A': 'nao',
                    'B': 'sim'}

df_enem['INTERNET'] = df_enem['INTERNET'].replace(replace_INTERNET)
```

```python
df_enem = pd.get_dummies(df_enem, columns=['TP_COR_RACA','TP_ESCOLA','TP_ENSINO','TP_DEPENDENCIA_ADM_ESC','TP_LOCALIZACAO_ESC','TP_SIT_FUNC_ESC','ESCOLARIDADE_PAI','ESCOLARIDADE_MAE','RENDA_MENSAL','INTERNET'])
```

```python
agg_methods = {'NU_NOTA_CN': 'mean',
                'NU_NOTA_CH': 'mean',
                'NU_NOTA_LC': 'mean',
                'NU_NOTA_MT': 'mean',
                'NU_NOTA_REDACAO': 'mean',
                'TP_COR_RACA_nao_declarado': 'sum',
                'TP_COR_RACA_branca': 'sum',
                'TP_COR_RACA_preta': 'sum',
                'TP_COR_RACA_parda': 'sum',
                'TP_COR_RACA_amarela': 'sum',
                'TP_COR_RACA_indigena': 'sum',
                'TP_ESCOLA_nao_respondeu': 'sum',
                'TP_ESCOLA_publica': 'sum',
                'TP_ESCOLA_privada':'sum',
                'TP_ENSINO_ensino_regular':'sum',
                'TP_ENSINO_educacao_especial':'sum',
                'TP_DEPENDENCIA_ADM_ESC_federal':'sum',
                'TP_DEPENDENCIA_ADM_ESC_estadual':'sum',
                'TP_DEPENDENCIA_ADM_ESC_municipal':'sum',
                'TP_DEPENDENCIA_ADM_ESC_privada':'sum',
                'TP_LOCALIZACAO_ESC_urbana':'sum',
                'TP_LOCALIZACAO_ESC_rural':'sum',
                'TP_SIT_FUNC_ESC_em_atividade':'sum',
                'TP_SIT_FUNC_ESC_paralisada':'sum',
                'TP_SIT_FUNC_ESC_extinta':'sum',
                'TP_SIT_FUNC_ESC_extinta_em_anos_anteriores':'sum',
                'ESCOLARIDADE_PAI_A':'sum',
                'ESCOLARIDADE_PAI_B':'sum',
                'ESCOLARIDADE_PAI_C':'sum',
                'ESCOLARIDADE_PAI_D':'sum',
                'ESCOLARIDADE_PAI_E':'sum',
                'ESCOLARIDADE_PAI_F':'sum',
                'ESCOLARIDADE_PAI_G':'sum',
                'ESCOLARIDADE_PAI_H':'sum',
                'ESCOLARIDADE_MAE_A':'sum',
                'ESCOLARIDADE_MAE_B':'sum',
                'ESCOLARIDADE_MAE_C':'sum',
                'ESCOLARIDADE_MAE_D':'sum',
                'ESCOLARIDADE_MAE_E':'sum',
                'ESCOLARIDADE_MAE_F':'sum',
                'ESCOLARIDADE_MAE_G':'sum',
                'ESCOLARIDADE_MAE_H':'sum',
                'RENDA_MENSAL_A':'sum',
                'RENDA_MENSAL_B':'sum',
                'RENDA_MENSAL_C':'sum',
                'RENDA_MENSAL_D':'sum',
                'RENDA_MENSAL_E':'sum',
                'RENDA_MENSAL_F':'sum',
                'RENDA_MENSAL_G':'sum',
                'RENDA_MENSAL_H':'sum',
                'RENDA_MENSAL_I':'sum',
                'RENDA_MENSAL_J':'sum',
                'RENDA_MENSAL_K':'sum',
                'RENDA_MENSAL_L':'sum',
                'RENDA_MENSAL_M':'sum',
                'RENDA_MENSAL_N':'sum',
                'RENDA_MENSAL_O':'sum',
                'RENDA_MENSAL_P':'sum',
                'RENDA_MENSAL_Q':'sum',
                'INTERNET_nao':'sum',
                'INTERNET_sim':'sum'
                }

df_enem = df_enem.groupby(["NU_ANO", "SG_UF_ESC", "NO_MUNICIPIO_ESC"]).agg(agg_methods).reset_index()
```

```python
df_enem.to_csv(os.getenv('PATH_TO_PROCESSED_DATA')+'enem_grouped_'+year+'.csv', index=False)
```

# Junção dos dados do Enem com os dados das Metas

```python
df_enem_2019 = pd.read_csv(os.getenv('PATH_TO_PROCESSED_DATA')+'enem_grouped_2019.csv')
df_enem_2020 = pd.read_csv(os.getenv('PATH_TO_PROCESSED_DATA')+'enem_grouped_2020.csv')
df_enem_2021 = pd.read_csv(os.getenv('PATH_TO_PROCESSED_DATA')+'enem_grouped_2021.csv')
```

```python
df_enem = pd.concat([df_enem_2019,df_enem_2020,df_enem_2021], axis=0, ignore_index=True)
df_enem.head()
```

```python
df_enem.columns
```

```python
df_enem
```

```python
df_enem = df_enem.groupby(['SG_UF_ESC', 'NO_MUNICIPIO_ESC']).mean().reset_index()
df_enem.head()
```

```python
import json
 
with open(os.getenv('PATH_TO_RAW_DATA')+'sigla_estado.json') as json_file:
    dict_sigla = json.load(json_file)
```

```python
df_enem['ESTADO'] = df_enem['SG_UF_ESC'].apply(lambda x: dict_sigla[x])
```

```python
df_municipios = pd.read_csv(os.getenv('PATH_TO_PROCESSED_DATA')+'municipios.csv')
df_municipios.head()
```

```python
df_meta = pd.read_csv(os.getenv('PATH_TO_RAW_DATA')+'FCT_META_TRES.csv')
df_meta
```

```python
df_meta = df_meta.merge(df_municipios, left_on='FK_MUNICIPIO_CODIGO', right_on='cod_municipio')
```

```python
df_meta = df_meta[(df_meta['ANO'] == 2021) | (df_meta['ANO'] == 2020) | (df_meta['ANO'] == 2019)]
```

```python
df_meta_enem = df_meta.merge(df_enem, left_on=['estado', 'municipio', 'ANO'], right_on=['ESTADO', 'NO_MUNICIPIO_ESC', 'NU_ANO'])
df_meta_enem.columns
```

```python
df_meta_enem['ENEM_NOTA'] = (df_meta_enem['NU_NOTA_CN'] + df_meta_enem['NU_NOTA_CH'] + df_meta_enem['NU_NOTA_LC'] + df_meta_enem['NU_NOTA_MT'] + df_meta_enem['NU_NOTA_REDACAO'])/5
```

```python
df_meta_enem.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df_meta_enem.dropna(subset=['ENEM_NOTA', 'PP_ATENDIMENTO_IND'], how='any', inplace=True)
```

```python
df_meta_enem.to_csv(os.getenv('PATH_TO_PROCESSED_DATA')+'meta_enem.csv', index=False)
```
